```python
import cv2
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8n.pt")
```

```python
ROAD_USERS = ["person", "bicycle", "car", "bus", "truck", "train"]

def detect_objects_on_road(model, image_path):
    """
    Detecta todos los objetos en una imagen.
    Retorna una lista de diccionarios con label, confianza y box_points.
    """
    results = model(image_path)
    detections = []

    for r in results:
        for box in r.boxes:
            cls_id = int(box.cls[0].item())
            label = model.names[cls_id]
            conf = float(box.conf[0].item())
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

            detections.append({
                "label": label,
                "confidence": conf,
                "box_points": (x1, y1, x2, y2)
            })

    return detections


def analyze_objects(detections, image_path, output_folder="detected_objects"):
    """
    Filtra solo usuarios de la carretera y guarda los crops.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image = cv2.imread(image_path)
    road_users = []

    for i, det in enumerate(detections):
        if det["label"] in ROAD_USERS:
            road_users.append(det)


            x1, y1, x2, y2 = det["box_points"]
            crop = image[y1:y2, x1:x2]

            filename = f"{det['label']}_{i}.jpg"
            filepath = os.path.join(output_folder, filename)
            cv2.imwrite(filepath, crop)

    return road_users


def draw_boxes(image_path, detections, save_path="output.jpg"):
    """
    Dibuja los recuadros de los objetos detectados en la imagen.
    """
    image = cv2.imread(image_path)

    for det in detections:
        x1, y1, x2, y2 = det["box_points"]
        label = det["label"]
        conf = det["confidence"]


        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)


        text = f"{label} {conf:.2f}"
        cv2.putText(image, text, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cv2.imwrite(save_path, image)


    img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()



test_image = "image.png"

detections = detect_objects_on_road(model, test_image)
print("Detecciones totales:", detections)

road_users = analyze_objects(detections, test_image)
print("Usuarios de la carretera detectados:", road_users)


draw_boxes(test_image, detections, save_path="output.jpg")
```
